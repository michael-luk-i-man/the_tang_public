```python
import pandas as pd, seaborn as sb, matplotlib.pyplot as mpl, plotly as pl, numpy as np
from statsmodels.graphics.mosaicplot import mosaic as mos
```

```python
# Here we will visualize the average distribution of sales in terms of items sold per day of the week. Each vertical column is a day of the week. The space each block takes up in a column represents the relative demand of that item compared to the others in that day. I will also label total items sold per day.
```

```python
wk_0 = '../the_tang/week_0_items.csv'

df = pd.read_csv(str(wk_0))
```

```python
display(df)
```

```python
# Iterate through day, manually due to nature of data:

item_dictionary = []
line = 0
for index, row in df.iterrows():
    item_dictionary.append([row[1],[]])
    for i in range(0,7):
        col = 7 + (i*2)
        item_dictionary[line][1].append(row[col])
    line += 1
```

```python
item_dictionary.pop(0)
display(item_dictionary)
```

```python
days_data = ({})
totals = []

weekdays = []

for i in range(0,7):
    weekdays.append([])
    for item in item_dictionary:
            if not isinstance(item[1][i], str):
                item[1][i] = 0
#             for j in range(i,int(item[1][i])):
            weekdays[i].append(item[1][i])


dict = { 'Mon': weekdays[0],
        'Tue': weekdays[1],
        'Wed': weekdays[2],
        'Thu': weekdays[3],
        'Fri': weekdays[4],
        'Sat': weekdays[5],
        'Sun': weekdays[6],
       }

display(dict)
data = pandas.DataFrame(dict)
display(data)
    
```
