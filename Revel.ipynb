```python
import csv, pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sn
```

```python
revel_data = pd.read_csv('sales_week_0.csv')
pd.set_option('display.max_columns', None)
```

```python
display(revel_data)
```

```python
def day_credit_to_tips(cred_tot, cred_tips):
    if cred_tot == 0:
        return 0
    perc = cred_tips/cred_tot
    return perc

for index, row in revel_data.iterrows():
    percent = day_credit_to_tips(row['Credit Total'], row['Credit Tips Total'])
    if row['Credit Total'] != 0:
        print(row['Time From'],'to',row['Time To'])
        print(row['Credit Total'],'Tips', row['Credit Tips Total'])
        print(percent, '\n')
```

```python
for index, row in revel_data.iterrows():
    cash_ratio = row['Cash Total']/row['Credit Total']
    if row['Credit Total'] != 0:
        print(row['Time From'],'to',row['Time To'])
        print(cash_ratio,'\n')
```
